```python
def plot_det(ts, te, ts_ev, df_gr_mask, df_ns_gr_mask, all_class, save_dir, plot_ending):
    
    '''
    ts: debris flow start times [list]
    te: debris flow end times [list]
    ts_ev: actual debris flow start time [string]
    df_gr_mask: slope failure detections summed hourly [DataFrame]
    df_ns_gr_mask: percipitation summed hourly [DataFrame]
    all_class: slope failure detections in time window NOT summed [Serie]
    save_dir: save direction [string]
    plot_ending: plot ending [string]
    '''
    
    fig_det, (ax1,ax0) = plt.subplots(2,1,figsize=(6.4, 4.8), sharex=True)
    cmcolors = plt.get_cmap('magma').colors
    colors = [cmcolors[20],cmcolors[120],cmcolors[180],cmcolors[200]]
    
    ax1.plot_date(df_ns_gr_mask.index, df_ns_gr_mask['261'], '-', color=colors[0], label='Air temperature at surface')
    ax1.plot(dttimes, [x*10 for x in all_class], 'o', alpha=0.3, color=colors[2], label='Slope failures')
    ax1.set_ylim(5,32)
    ax1.set_xlim(df_gr_mask.index[1],df_gr_mask.index[-2])
    ax1.tick_params(axis='x', rotation=25)

    ax2 = ax1.twinx()
    ax2.plot_date(df_ns_gr_mask.index, df_ns_gr_mask['306'], '-',color=colors[3], label='Atmospheric pressure')
    ax1.set_ylabel('Temperature (°C)')
    ax2.set_ylabel('Pressure (hPa)')
    ax2.set_ylim(840,880)

    # Plot bar plot with slope failures per hour
    ax0.bar(df_gr_mask.index, df_gr_mask['classes'], width=0.042, color=colors[2],  align='edge', label='Nb of slope failures \nper hour')
    ax0.xaxis_date()
    ax0.set_ylabel('Slope failure \nwindows')
    ax0.tick_params(axis='x', rotation=25)
    ymax = np.max(df_gr_mask['classes'])+0.25*np.max(df_gr_mask['classes'])
    ax0.set_ylim(0,ymax)

    # Plot bar plot with precipitation per hour
    if df_ns_gr_mask['267'].sum()!=0:
        ax3 = ax0.twinx()
        ax3.bar(df_ns_gr_mask.index, df_ns_gr_mask['267'], width=0.042 ,color=colors[0], label='Precipitation per hour', align='edge')
        ax3.xaxis_date()
        ax3.set_ylabel('Precipitation \n(mm)')
        ax3.set_ylim(0,np.max(df_ns_gr_mask['267'])/(ymax-np.max(df_gr_mask['classes']))*ymax) # adjust the second yscale to 100%
        ax3.set_ylim(ax3.get_ylim()[::-1]) # slip y scale

    #plt.legend(bbox_to_anchor = (1.04,1), loc="upper left")

    # Plot debris-flow events as vertical lines
    for ts_i, te_i in zip(ts,te):
        ax1.axvline(datetime.strptime(ts_i,"%Y-%m-%dT%H:%M:%S"), linestyle='-', color=colors[1])
        ax1.axvline(datetime.strptime(te_i,"%Y-%m-%dT%H:%M:%S"), linestyle=':', color=colors[1])
        ax0.axvline(datetime.strptime(ts_i,"%Y-%m-%dT%H:%M:%S"), linestyle='-', color=colors[1])
        ax0.axvline(datetime.strptime(te_i,"%Y-%m-%dT%H:%M:%S"), linestyle=':', color=colors[1])

    t_name = ts_ev.replace(':','.') # replace : with . for figure name
    plt.savefig(save_dir + f'/{t_name}_{plot_ending}.png',bbox_inches='tight', dpi=300)
    return(fig_det)
```

```python
def calc_spec(year, ts_ev, tds, tde, path_to_raw_data):
    
    '''
    year: year [integer]
    ts_ev: actual debris flow start time [string]
    tds: 7 h before debris flow starts [datetime]
    tde: 7 h after debris flow ends [datetime]
    path_to_raw_data: path to raw seismic data [string]
    '''
    
    # Get raw data
    st = obspy.read(path_to_raw_data + f"{year}/XP/ILL?8/EHZ.D/*.ILL?8.*{tds.strftime('%j')}") # julian day

    # Append raw data if endtime is on the next day 
    if tde > st[0].stats.endtime:
        st1 = obspy.read(path_to_raw_data + f"{year}/XP/ILL?8/EHZ.D/*.ILL?8.*{int(tds.strftime('%j')) + 1}")
        for tr in st1:
            st += tr

    # Some preprocessing
    st.merge(fill_value='interpolate')
    st.trim(obspy.UTCDateTime(tds), obspy.UTCDateTime(tde))
    st.merge(method='interpolate')
    st.taper(0.01)
    tr = st[0]

    # Compute spectrogram
    f, t, Sxx = signal.spectrogram(tr.data, tr.stats.sampling_rate, nperseg=1024, noverlap=512, nfft=2048)
    tr1 = tr.copy().filter('bandpass', freqmin=5, freqmax=10)
    return(tr, tr1, f, t, Sxx)
```

```python
def plot_spec(tr, tr1, t, Sxx, tslops, dslops,
              ts, te, ts_ev, tds, tde, df_gr_mask, df_ns_gr_mask, save_dir, plot_ending):      
    
    '''
    tr: trace
    tr1: filtered trace
    t: spectrogram bin times [array 1D]
    Sxx: calculated spectrogram [array 2D]
    tslops: timestamp of slope failure detections [list]
    dslops: 1 same length as tslops [list]
    ts: debris flow start times [list]
    te: debris flow end times [list]
    tds: 7 h before debris flow starts [datetime]
    tde : 7 h after debris flow ends [datetime]
    df_gr_mask: slope failure detections summed hourly [DataFrame]
    df_ns_gr_mask: percipitation summed hourly [DataFrame]
    save_dir: save direction [string]
    plot_ending: plot ending [string]
    '''
    
    fig_spec, (ax1,ax2,ax0) = plt.subplots(3,1,figsize=(6.4, 4.8), sharex=True)
    cmcolors = plt.get_cmap('magma').colors
    cmap = plt.get_cmap('magma')
    # Specify colors
    colors = [cmcolors[20],cmcolors[120],cmcolors[180],cmcolors[200]]
    #ax1.plot_date(ns_dat.index, ns_dat['261'], '-', color=colors[0], label='Air temperature')

    # Plot raw data
    ax1.plot_date(tr1.times('matplotlib'), tr1.data,'k-', linewidth=0.1, label=tr.id)

    # Plot spectrogram
    im = ax2.pcolormesh([tr.stats.starttime.datetime + timedelta(seconds=x) for x in t], f, 10*np.log10(Sxx), cmap=cmap, vmin=2,vmax=25,rasterized=True)

    # Plot slope failures
    ax1.plot(tslops, dslops, 'o', alpha=0.3, color=colors[2], label='Slope failures')

    ax2.set_ylabel('Frequency (Hz)')
    ax1.set_ylabel('Amplitude')

    # Plot bar plot with slope failures per hour
    ax0.bar(df_gr_mask.index, df_gr_mask['classes'], width=0.042, color=colors[2],  align='edge', label='Nb of slope failures \nper hour')
    ax0.xaxis_date()
    ax0.set_ylabel('Slope failure \nwindows')
    ax0.tick_params(axis='x', rotation=25)
    ymax = np.max(df_gr_mask['classes'])+0.25*np.max(df_gr_mask['classes'])
    ax0.set_ylim(0,ymax)

    # Plot bar plot with precipitation per hour
    ax3 = ax0.twinx()
    ax3.bar(df_ns_gr_mask.index, df_ns_gr_mask['267'], width=0.042 ,color=colors[0], label='Precipitation per hour', align='edge')
    ax3.xaxis_date()
    ax3.set_ylabel('Precipitation \n(mm)')
    ax3.set_ylim(0,np.max(df_ns_gr_mask['267'])/(ymax-np.max(df_gr_mask['classes']))*ymax) # adjust the second yscale to 100%
    ax3.set_ylim(ax3.get_ylim()[::-1]) # slip y scale

    # Plot debris-flow start and end
    for ax in (ax3,ax1,ax2):

        # Plot debris-flow events as vertical lines
        for ts_i, te_i in zip(ts,te):
            if ax == ax1:
                ax.axvline(datetime.strptime(ts_i,"%Y-%m-%dT%H:%M:%S"), color=colors[1], label= 'Debris flow start')
                ax.axvline(datetime.strptime(te_i,"%Y-%m-%dT%H:%M:%S"), color=colors[1], linestyle=':',label= 'Debris flow end')
            elif ax == ax2:
                ax.axvline(datetime.strptime(ts_i,"%Y-%m-%dT%H:%M:%S"), color='white')
                ax.axvline(datetime.strptime(te_i,"%Y-%m-%dT%H:%M:%S"), color='white', linestyle=':')
            else:
                ax.axvline(datetime.strptime(ts_i,"%Y-%m-%dT%H:%M:%S"), color=colors[1])
                ax.axvline(datetime.strptime(te_i,"%Y-%m-%dT%H:%M:%S"), color=colors[1], linestyle=':')

    ax0.set_xlim(tds + timedelta(seconds=1*3600), tde - timedelta(seconds=1*3600))

    #ax3.legend(loc="upper right")
    #ax0.legend(loc='lower right')
    fig_spec.legend(bbox_to_anchor = [1, 0.75])
    #ax2.legend(loc='lower right')
    annot = ['(a)','(b)', '(c)']
    axes = (ax1, ax2, ax0)
    for ax, an in zip(axes,annot):
        xlim = ax.get_xlim()
        ylim = ax.get_ylim()
        if an == '(b)':
            ax.text(xlim[0] + 0.02*(xlim[1] - xlim[0]), ylim[1] - 0.25*(ylim[1] - ylim[0]), an, color='white')
        elif an == '(a)':
            ax.text(xlim[0] + 0.02*(xlim[1] - xlim[0]), ylim[1] - 0.25*(ylim[1] - ylim[0]), an)
            ax.text(xlim[0] + 0.45*(xlim[1] - xlim[0]), ylim[1] - 0.25*(ylim[1] - ylim[0]), tr.stats.starttime.ctime())
        else:
            ax.text(xlim[0] + 0.02*(xlim[1] - xlim[0]), ylim[1] - 0.25*(ylim[1] - ylim[0]), an)

    ax1.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
    ax0.ticklabel_format(axis='y', style='sci', scilimits=(0,0))

    t_name = ts_ev.replace(':','.') # replace : with . for figure name
    plt.savefig(save_dir + f'/0_eval_{t_name}_{plot_ending}.png',bbox_inches='tight', dpi=300)
    return(fig_spec)
```

```python
def plot_all(tr, tr1, f, t, Sxx, tslops, dslops,
              ts, te, ts_ev, tds, tde, df_gr_mask, df_ns_gr_mask, all_class, save_dir, plot_ending):      
    
    '''
    tr: trace
    tr1: filtered trace
    f: 
    t: spectrogram bin times [array 1D]
    Sxx: calculated spectrogram [array 2D]
    tslops: timestamp of slope failure detections [list]
    dslops: 1 same length as tslops [list]
    ts: debris flow start times [list]
    te: debris flow end times [list]
    tds: 7 h before debris flow starts [datetime]
    tde : 7 h after debris flow ends [datetime]
    df_gr_mask: slope failure detections summed hourly [DataFrame]
    df_ns_gr_mask: percipitation summed hourly [DataFrame]
    all_class: slope failure detections in time window NOT summed [Serie]
    save_dir: save direction [string]
    plot_ending: plot ending [string]
    '''
    
    fig_all, (ax1,ax2,ax4,ax0) = plt.subplots(4,1,figsize=(6.4, 4.8), sharex=True)
    cmcolors = plt.get_cmap('magma').colors
    cmap = plt.get_cmap('magma')
    # Specify colors
    colors = [cmcolors[20],cmcolors[120],cmcolors[180],cmcolors[200]]
    #ax1.plot_date(ns_dat.index, ns_dat['261'], '-', color=colors[0], label='Air temperature')

#--------------------------------------------------------------------------------------------------------------    
    ax1.plot_date(tr1.times('matplotlib'), tr1.data,'k-', linewidth=0.1, label=tr.id) # Plot raw data
    ax1.plot(tslops, dslops, 'o', alpha=0.3, color=colors[2], label='Slope failures') # Plot slope failures
    ax1.set_ylabel('Amplitude')
#--------------------------------------------------------------------------------------------------------------     
    # Plot spectrogram
    im = ax2.pcolormesh([tr.stats.starttime.datetime + timedelta(seconds=x) for x in t], f, 10*np.log10(Sxx), cmap=cmap, vmin=2,vmax=25,rasterized=True)
    ax2.set_ylabel('Frequency (Hz)')
#-------------------------------------------------------------------------------------------------------------- 
    ax4.plot_date(df_ns_gr_mask.index, df_ns_gr_mask['261'], '-', color=colors[0], label='Air temperature at surface')
    ax4.plot(dttimes, [x*10 for x in all_class], 'o', alpha=0.3, color=colors[2])
    ax4.set_ylim(5,32)
    ax4.set_xlim(df_gr_mask.index[1],df_gr_mask.index[-2])
    ax4.tick_params(axis='x', rotation=25)

    ax5 = ax4.twinx()
    ax5.plot_date(df_ns_gr_mask.index, df_ns_gr_mask['306'], '-',color=colors[3], label='Atmospheric pressure')
    ax4.set_ylabel('Temperature (°C)')
    ax5.set_ylabel('Pressure (hPa)')
    ax5.set_ylim(840,880)
#-------------------------------------------------------------------------------------------------------------- 
    # Plot bar plot with slope failures per hour
    ax0.bar(df_gr_mask.index, df_gr_mask['classes'], width=0.042, color=colors[2],  align='edge', label='Nb of slope failures \nper hour')
    ax0.xaxis_date()
    ax0.set_ylabel('Slope failure \nwindows')
    ax0.tick_params(axis='x', rotation=25)
    ymax = np.max(df_gr_mask['classes'])+0.25*np.max(df_gr_mask['classes'])
    ax0.set_ylim(0,ymax)
    ax0.set_xlim(tds + timedelta(seconds=1*3600), tde - timedelta(seconds=1*3600))

    # Plot bar plot with precipitation per hour
    ax3 = ax0.twinx()
    ax3.bar(df_ns_gr_mask.index, df_ns_gr_mask['267'], width=0.042 ,color=colors[0], label='Precipitation per hour', align='edge')
    ax3.xaxis_date()
    ax3.set_ylabel('Precipitation \n(mm)')
    ax3.set_ylim(0,np.max(df_ns_gr_mask['267'])/(ymax-np.max(df_gr_mask['classes']))*ymax) # adjust the second yscale to 100%
    ax3.set_ylim(ax3.get_ylim()[::-1]) # slip y scale
#-------------------------------------------------------------------------------------------------------------- 
    # Plot debris-flow start and end
    for ax in (ax4,ax3,ax1,ax2):
        count = 0
        # Plot debris-flow events as vertical lines
        for ts_i, te_i in zip(ts,te):
            if ax == ax1 and count == 0:
                ax.axvline(datetime.strptime(ts_i,"%Y-%m-%dT%H:%M:%S"), color=colors[1], label= 'Debris flow start')
                ax.axvline(datetime.strptime(te_i,"%Y-%m-%dT%H:%M:%S"), color=colors[1], linestyle=':',label= 'Debris flow end')
                count += 1
            elif ax == ax2:
                ax.axvline(datetime.strptime(ts_i,"%Y-%m-%dT%H:%M:%S"), color='white')
                ax.axvline(datetime.strptime(te_i,"%Y-%m-%dT%H:%M:%S"), color='white', linestyle=':')
            else:
                ax.axvline(datetime.strptime(ts_i,"%Y-%m-%dT%H:%M:%S"), color=colors[1])
                ax.axvline(datetime.strptime(te_i,"%Y-%m-%dT%H:%M:%S"), color=colors[1], linestyle=':')

    #ax3.legend(loc="upper right")
    #ax0.legend(loc='lower right')
    fig_all.legend(bbox_to_anchor = [1.6, 0.75])
    #ax2.legend(loc='lower right')
    annot = ['(a)','(b)', '(c)', '(d)']
    axes = (ax1, ax2, ax0, ax4)
    for ax, an in zip(axes,annot):
        xlim = ax.get_xlim()
        ylim = ax.get_ylim()
        if an == '(b)':
            ax.text(xlim[0] + 0.02*(xlim[1] - xlim[0]), ylim[1] - 0.25*(ylim[1] - ylim[0]), an, color='white')
        elif an == '(a)':
            ax.text(xlim[0] + 0.02*(xlim[1] - xlim[0]), ylim[1] - 0.25*(ylim[1] - ylim[0]), an)
            ax.text(xlim[0] + 0.55*(xlim[1] - xlim[0]), ylim[1] - 0.25*(ylim[1] - ylim[0]), ts_ev)
        else:
            ax.text(xlim[0] + 0.02*(xlim[1] - xlim[0]), ylim[1] - 0.25*(ylim[1] - ylim[0]), an)

    ax1.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
    ax0.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
    
    t_name = ts_ev.replace(':','.') # replace : with . for figure name
    plt.savefig(save_dir + f'/all_{t_name}_{plot_ending}.png',bbox_inches='tight', dpi=300)
    return(fig_all)
```

```python
# loop over debris flow start time
def main(ts_ev, te_ev, df_group, ns_group, all_class, tslops, dslops,
         year, save_dir, plot_ending, path_to_raw_data):
    print(ts_ev)
    # Get start and endtime
    tstart = obspy.UTCDateTime(ts_ev)
    tend = obspy.UTCDateTime(te_ev)
    
    tds = tstart.datetime- timedelta(seconds=7*3600) # start 7 hours bevore event
    tde = tend.datetime + timedelta(seconds=7*3600) # end 7 hours after event

    gr_mask = (df_group.index > tds) & (df_group.index <= tde) # daily
    ns_gr_mask = (ns_group.index > tds) & (ns_group.index<= tde) #daily
    df_gr_mask = df_group[gr_mask]
    df_ns_gr_mask = ns_group[ns_gr_mask]
    
    if not df_gr_mask.empty: # if we have seismic data for the debris flow
        
        # plot slope failure detections around a debris flow and some weather parameter
#        plot_det(ts, te, ts_ev, df_gr_mask, df_ns_gr_mask, all_class, save_dir, plot_ending)

#---------------------------------------------------------------------------------------------------------------        
        # load stream(s) around debris flow, do some processing and calculate the spectrogram
        tr, tr1, f, t, Sxx = calc_spec(year, ts_ev, tds, tde, path_to_raw_data)
        
#---------------------------------------------------------------------------------------------------------------
        # plot amplitude, spectrogram, weather parameter and slope failure detections
        #plot_spec(tr,tr1,t,Sxx, tslops, dslops, df_gr_mask, df_ns_gr_mask, ts, te, ts_ev, tds, tde, save_dir, plot_ending)
#---------------------------------------------------------------------------------------------------------------
        # plot all in one
        plot_all(tr, tr1, f, t, Sxx, tslops, dslops,
              ts, te, ts_ev, tds, tde, df_gr_mask, df_ns_gr_mask, all_class, save_dir, plot_ending)
        print(f'Done: {ts_ev}')
    return
```

```python
# read df
year = 2021
file_ending = 'test_v4_bigsf_clf2'

class_res = pd.read_csv(f'../detections/{year}/{year}_{file_ending}.txt')
class_res['datetimes'] = pd.to_datetime(class_res['datetimes'])
dttimes = class_res['datetimes']
all_class = class_res['classes']

# Get times of all slope failure events
tslops = []
dslops = []
for x,y in zip(dttimes, all_class):
    if y != 0:
        tslops.append(x) # list with timestamps when detection
        dslops.append(y) # list with 1's for slope failure detections
len(dslops)

plot_ending = 'v4_bigsf_clf2'
path_to_raw_data = '/media/manuela/Elements/illgraben/miniseed/'
```

```python
# Group by day for barplot (noise class will be 0 still)
df_group_d = class_res.groupby([class_res['datetimes'].dt.date]).sum() # daily
df_group = class_res.resample('H', on='datetimes').sum() # hourly
df_group_d.head()
```

```python
# Get precipitation data (CransMontana) (ns short for the german Niederschlag (precipitation))
def read_MeteoSwiss(path, files='all', header=14):
    if files == 'all':
        lsmet = os.listdir(path)
    else:
        lsmet = files
        
    met = pd.DataFrame()
    for f in lsmet:
        filepath = os.path.join(path, f)
        d = pd.read_csv(filepath, header = header, delim_whitespace = True)
        d = d.assign(Date = pd.NaT)
        d.Date = pd.to_datetime(dict(year=d.JAHR, month=d.MO, day=d.TG, hour=d.HH, minute=d.MM)) # convert columns (dtype=object) to dates
        d = d.drop(['JAHR','MO','TG','HH','MM','STA'], axis=1) # drop cols, bcs they are in date now
        met = met.append(d)
    
    met = met.set_index('Date')
    
    return met

ns_dat = read_MeteoSwiss('../meteodata/', files=[f'{year}.dat'])
ns_dat['datetime'] = ns_dat.index
ns_dat.head()
```

```python
# Group by date/hour and sum
ns_group_d = ns_dat.groupby(ns_dat['datetime'].dt.date).sum() # daily
ns_group = ns_dat.resample('H', on='datetime').sum() # hourly
ns_group_d.head()
```

```python
ts,te = DF_detections(year) # start and endtimes of DFs
```
