# Read snuffler-file & change time-format

I use the first part of this Notebook to preparate the snuffler-files to files with start and end date and time, plus the event classs. The output is saved in plot_classes, filename snuffler_detections.

```python
import numpy as np
import pandas as pd
import datetime
from obspy import UTCDateTime
```

```python
# read snuffler marker file
path_to_raw_data = '/data/wsd03/data_manuela/Illgraben/miniseed/'
df = pd.read_csv('../snuffler_files/markers_end2017_snuffler_new.txt', 
                 delim_whitespace=True, skiprows=1 ,names=['sd', 'st', 'ed', 'et', 'dt', 'class', 'stat'])
df
```

```python
# test
datetime.datetime.strptime(df.sd.iloc[0], '%Y-%m-%d')<datetime.datetime.strptime('2017-5-30', '%Y-%m-%d')
```

```python
# creae array with wished dates as string --> date MUST be 1d smaller/larger den wished time period
date_l = [date for date in df['sd']
          if datetime.datetime.strptime(date, '%Y-%m-%d')>datetime.datetime.strptime('2017-5-28', '%Y-%m-%d') and 
          datetime.datetime.strptime(date, '%Y-%m-%d')<datetime.datetime.strptime('2017-6-8', '%Y-%m-%d')]
date_l = np.unique(date_l)
date_l
```

```python
# keep only wished dates
df = df[df['sd'].isin(date_l)]
df
```

## Prepare times to plot spectrogram

```python
df_det = df[df['stat']=='None'] # detected
df_det
```

```python
df_pik = df[df['stat']!='None'] # picked
df_pik
```

```python
# dfx = df_det.copy()
dfx = df_pik.copy()
```

```python
starttimes = [UTCDateTime("{}T{}".format(dfx['sd'].iloc[i],dfx['st'].iloc[i]),)-10 for i in range(len(dfx['st']))]
endtimes = [UTCDateTime("{}T{}".format(dfx['ed'].iloc[i],dfx['et'].iloc[i]))+30 for i in range(len(dfx['et']))]
classes = dfx['class'].to_list()
```

```python
df_new = pd.DataFrame()
df_new['starttimes'] = starttimes
df_new['endtimes'] = endtimes
df_new['classes'] = classes
df_new
```

```python
df_new.to_csv('../plot_classes/snuffler_detections_end2017_new.csv', index=False)
```

# Separate different event types

The event_catalog is splited into all the different classes and saved as a separate file called event_catalog ...

```python
import numpy as np
import pandas as pd
```

```python
# read file
# df = pd.read_csv('../snuffler_files/picks_for_snuffler_hSNR_v4.txt', 
#                  delim_whitespace=True, skiprows=1 ,names=['sd', 'st', 'ed', 'et', 'dt', 'class', 'stat'])
df = pd.read_csv('../snuffler_files/event_catalog_man_cont_hSNR_v4.txt')
df
```

```python
idx = np.where(df.network.map(type)!=str)[0] # where we only have start time
# shift class and network to right position at places where we only have start time
for i in idx:
    df.loc[i,'class'] = float(df.loc[i,'endtime']) # entery form endtime to class
    df.loc[i,'endtime'] = np.nan # enttime to NaN
    df.loc[i,'network'] = df.loc[i]['duration'] # entery form duration to network
    df.loc[i,'duration'] = np.nan # duration to NaN
```

```python
np.unique(df['class'])
```

```python
df_eq = df.loc[df['class'] == 1.0]
df_eq
```

```python
df_eq.to_csv('../snuffler_files/event_catalog_eq_v4.txt', sep=',', index=False)
```

```python
df_noise = df.loc[df['class'] == 0.0]
df_noise
```

```python
df_noise.to_csv('../snuffler_files/event_catalog_noise_v4.txt', sep=',', index=False)
```

```python
df_sf = df.loc[df['class'] == 3.0]
df_sf
```

```python
df_sf.to_csv('../snuffler_files/event_catalog_sf_v4.txt', sep=',', index=False)
```

```python
df_ev_noise = df.loc[df['class'] == 4.0]
df_ev_noise
```

```python
df_ev_noise.to_csv('../snuffler_files/event_catalog_ev_noise_v4.txt', sep=',', index=False)
```

```python
# read snuffler marker file
df_thunder = pd.read_csv('../snuffler_files/picks_for_snuffler_thunder_v4.txt', 
                 delim_whitespace=True, skiprows=1 ,names=['startdate','starttime','enddate','endtime','duration',
                                                           'class','network'])
df_thunder.insert(loc = 5, column = 'bla', value = np.nan)
df_thunder['class'] = df_thunder['class'].astype('float')
df_thunder
```

```python
df_thunder.to_csv('../snuffler_files/event_catalog_thunder_v4.txt', sep=',', index=False)
```

# Merge different event types new together

This part merges different event types new together and create input file for cascadua_make_feature_files.ipynb ich is used to create the training and testing data set.

```python
import pandas as pd
```

```python
df_sf = pd.read_csv('../snuffler_files/event_catalog_sf_v4.txt')

df_noise = pd.read_csv('../snuffler_files/event_catalog_noise_v4.txt')
df_ev_noise = pd.read_csv('../snuffler_files/event_catalog_ev_noise_v4.txt')
df_thunder = pd.read_csv('../snuffler_files/event_catalog_thunder_v4.txt')
df_eq = pd.read_csv('../snuffler_files/event_catalog_eq_v4.txt')
```

```python
df_final = pd.concat([df_noise,df_ev_noise,df_thunder,df_eq, df_sf])
df_final
```

```python
df_final.to_csv('../snuffler_files/catalog_beginning_2017.txt', sep=',', index=False)
```
