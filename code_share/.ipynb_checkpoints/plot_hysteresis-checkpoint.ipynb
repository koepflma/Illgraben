```python
import numpy as np
import pandas as pd
import obspy
import matplotlib
import matplotlib.pyplot as plt
import os
from datetime import datetime, timedelta
import time
from scipy import signal
from ComputeAttributes_CH_V1 import *
from DF_detections import *
from ILL_functions import *
```

```python
# set plotting default parameters
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['figure.figsize'] = (6.4, 4.8) # if wider plot, only change first value.
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Verdana']
#plt.rcParams['ytick.major.pad']= 2

cmcolors = plt.get_cmap('viridis').colors #magma or viridis
cmap = plt.get_cmap('viridis')
# Specify colors
colors = [cmcolors[20],cmcolors[120],cmcolors[180],cmcolors[200]]
```

```python
# read df
year = 2020
file_ending = 'test_v4_bigsf_clf2'
plot_ending = 'v4_bigsf_clf2'
path_to_raw_data = '/media/manuela/Elements/illgraben/miniseed/'

class_res, dttimes, all_class, tslops, dslops = read_features(year, file_ending)
df_group_d, df_group, ns_dat, ns_group_d, ns_group = read_meteo(class_res, year)

ts,te = DF_detections(year)
```

```python
# with Moisture
class_mask, ns_dat_per, ns_dat_mask = mask_debrisflow(class_res, ns_dat, ts, te)
df_mois_per,df_mois_mask = mask_moisture(class_res, ts, te)
df_polar, df_polar_mask, ns_polar, ns_polar_mask = mask_polar(class_res, class_mask, ns_dat, ns_dat_per, ns_dat_mask)
group_all_DF, group_all, group_all5_DF, group_all5 = some_processing_with_moisture(class_res, class_mask, ns_dat_per, ns_dat_mask, df_mois_per, df_mois_mask)
```

```python
group_all_DF_plot = group_all_DF.copy()
group_all_DF_plot['267'].loc[group_all_DF_plot['267']== 0.0] = 0.06 # replace percipitation 0.0with 0.06
group_all_DF_plot # with this df all points can be plottet in a loglog plot
```

```python
i_plot_ending = ['', '_withoutDF', '_6h', '_6h_withoutDF']
i_axlabel = ['', '', '6', '6']
df_list = [group_all_DF, group_all, group_all5_DF, group_all5]

for i_plt_ending, i_ax, df in zip(i_plot_ending, i_axlabel, df_list):
    N_TICKS = 10
    fig, ax = plt.subplots()
    smap = ax.scatter(df['SM10cm'],df['classes'],c=df.index, alpha=0.8,
                      edgecolors='k', marker='o', cmap=cmap)
    #plt.xlabel(f'Percipitation (mm/{i_ax}h)')
    plt.xlabel(f'Moisture (saturation/{i_ax}h)')
    plt.ylabel(f'Slope failures (detections/{i_ax}h)')
    ax.set_xscale('log')
    ax.set_yscale('log')
    #plt.xlim(0.05, max(df['267'])+max(df['267'])/5)
    plt.xlim(min(df['SM10cm'])-min(df['SM10cm'])/5, max(df['SM10cm'])+max(df['SM10cm'])/5)
    plt.ylim(0.8,max(df['classes'])+int(max(df['classes'])/5))
    indexes = [df.index[i] for i in np.linspace(0,df.shape[0]-1,N_TICKS).astype(int)] 

    cb = fig.colorbar(smap, orientation='vertical',
                      ticks= df.loc[indexes].index.astype(int))

    cb.ax.set_yticklabels([index.strftime('%d %b') for index in indexes])

    fig.text(0.15,0.83, f'{year}')
    plt.savefig(f'../plot/{year}/evolution_log_moisture10cm{i_plt_ending}_{plot_ending}.png',bbox_inches='tight', dpi=300)
```

```python
from matplotlib.offsetbox import AnchoredText

i_plot_ending = ['', '_withoutDF', '_6h', '_6h_withoutDF']
i_axlabel = ['', '', '6', '6']
df_list = [group_all_DF, group_all, group_all5_DF, group_all5]
ii = [0, 1, 0, 1]
jj = [0, 0, 1, 1]

fig, ax = plt.subplots(2,2,figsize=(6.4*2, 4.8*2), sharex=True)
for i_plt_ending, i_ax, df, i, j in zip(i_plot_ending, i_axlabel, df_list, ii, jj):
    N_TICKS = 10

    smap = ax[i,j].scatter(df['SM30cm'],df['classes'],c=df.index, alpha=1,
                      edgecolors='k', marker='o', cmap=cmap)
    if i == 1:
        #ax[i,j].set_xlabel(f'Percipitation (mm/{i_ax}h)')
        ax[i,j].set_xlabel(f'Moisture (saturaion/{i_ax}h)')
    ax[i,j].set_ylabel(f'Slope failures (detections/{i_ax}h)')
    ax[i,j].set_xscale('log')
    ax[i,j].set_yscale('log')
    #ax[i,j].set_xlim(0.08,max(df['267'])+int(max(df['267'])/5))
    ax[i,j].set_xlim(0.008,max(df['SM30cm'])+max(df['SM30cm'])/2)
    ax[i,j].set_ylim(0.8,max(df['classes'])+max(df['classes'])/5)

    #ax[i,j].add_artist(AnchoredText(f'{year}', loc='upper right'))

#fig.subplots_adjust(right=0.5)
cb_ax = fig.add_axes([1, 0.15, 0.03, 0.7]) # [left bottom width height] 

indexes = [df.index[idx] for idx in np.linspace(0,df.shape[0]-1,N_TICKS).astype(int)] 

cb = fig.colorbar(smap, orientation='vertical', label = f'{year}', ticks= df.loc[indexes].index.astype(int), cax=cb_ax)

cb.ax.set_yticklabels([index.strftime('%d %b') for index in indexes])

fig.tight_layout()
#plt.savefig(f'../plot/{year}/evolution_log_{plot_ending}.png',bbox_inches='tight', dpi=300)
```
